```python
import pandas as pd

# Load the CSVs
india_df = pd.read_csv('INvideos.csv')
usa_df = pd.read_csv('USvideos.csv')

# Preview the first few rows
print("India Data:")
display(india_df.head())

print("\nUSA Data:")
display(usa_df.head())
```

```python
# Add a 'Country' column to each DataFrame
india_df['Country'] = 'India'
usa_df['Country'] = 'USA'

# Convert 'publishedAt' to datetime if it exists (else use 'publish_time')
if 'publishedAt' in india_df.columns:
    india_df['publish_time'] = pd.to_datetime(india_df['publishedAt'], errors='coerce')
    usa_df['publish_time'] = pd.to_datetime(usa_df['publishedAt'], errors='coerce')
else:
    india_df['publish_time'] = pd.to_datetime(india_df['publish_time'], errors='coerce')
    usa_df['publish_time'] = pd.to_datetime(usa_df['publish_time'], errors='coerce')

# Drop rows with missing publish_time
india_df.dropna(subset=['publish_time'], inplace=True)
usa_df.dropna(subset=['publish_time'], inplace=True)

# Optional: Reset index
india_df.reset_index(drop=True, inplace=True)
usa_df.reset_index(drop=True, inplace=True)

# Confirm shapes
print(f"India Data Shape: {india_df.shape}")
print(f"USA Data Shape: {usa_df.shape}")
```

```python
# Check null values in India data
print("India missing values:")
print(india_df.isnull().sum())

print("\nUSA missing values:")
print(usa_df.isnull().sum())
```

```python
# Check null values in India data
print("India missing values:")
print(india_df.isnull().sum())

print("\nUSA missing values:")
print(usa_df.isnull().sum())
```

```python
# Combine India and USA datasets into one
combined_df = pd.concat([india_df, usa_df], ignore_index=True)

# Preview the combined dataset
print("Combined Data Shape:", combined_df.shape)
combined_df[['title', 'category_id', 'publish_time', 'Country']].head()
```

```python
import json

# Load the JSON file
with open('US_category_id.json') as f:
    data = json.load(f)

# Create mapping from ID to name
category_map = {}
for item in data['items']:
    category_map[int(item['id'])] = item['snippet']['title']

# Map in the combined DataFrame
combined_df['Category'] = combined_df['category_id'].map(category_map)

# Check if it worked
combined_df[['category_id', 'Category', 'Country']].head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set chart style
sns.set(style="whitegrid")

# Group by category and country
category_counts = combined_df.groupby(['Country', 'Category']).size().reset_index(name='Count')

# Create barplot
plt.figure(figsize=(14, 7))
sns.barplot(data=category_counts, x='Category', y='Count', hue='Country', palette='Set2')

plt.title('Top Trending Categories in India vs USA', fontsize=16)
plt.xlabel('Category')
plt.ylabel('Number of Trending Videos')
plt.yscale('log')

plt.xticks(rotation=70)
plt.legend(title='Country')
plt.yscale('log')

plt.tight_layout()
plt.show()
```

```python
# Create a new column that stores the length of each title
combined_df['title_length'] = combined_df['title'].apply(len)

# Calculate average title length for each country
title_length_avg = combined_df.groupby('Country')['title_length'].mean().reset_index()

# Plot the average title length
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.barplot(data=title_length_avg, x='Country', y='title_length', palette='pastel')

plt.title('Average Title Length by Country', fontsize=14)
plt.ylabel('Avg. Title Length (characters)')
plt.xlabel('Country')
plt.tight_layout()
plt.show()
```

```python
# Create a new column that checks if title is ALL CAPS
combined_df['all_caps'] = combined_df['title'].apply(lambda x: x.isupper())

# Count how many all-caps titles each country has
caps_count = combined_df.groupby('Country')['all_caps'].sum().reset_index()

# Plot it
plt.figure(figsize=(6, 4))
sns.barplot(data=caps_count, x='Country', y='all_caps', palette='cool')

plt.title('ALL CAPS Titles by Country')
plt.ylabel('Count of ALL CAPS Titles')
plt.xlabel('Country')
plt.tight_layout()
plt.show()
```

```python
import re

# Function to check if a title has emojis
def contains_emoji(text):
    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags
        u"\U00002700-\U000027BF"  # dingbats
        u"\U000024C2-\U0001F251"
        "]+", flags=re.UNICODE)
    return bool(emoji_pattern.search(text))

# Add a new column
combined_df['has_emoji'] = combined_df['title'].apply(contains_emoji)

# Group by country
emoji_count = combined_df.groupby('Country')['has_emoji'].sum().reset_index()

# Plot the result
plt.figure(figsize=(6, 4))
sns.barplot(data=emoji_count, x='Country', y='has_emoji', palette='flare')

plt.title('Emoji Use in Video Titles by Country')
plt.ylabel('Count of Titles with Emojis')
plt.xlabel('Country')
plt.tight_layout()
plt.show()
```

```python
# Extract the hour from publish_time
combined_df['publish_hour'] = combined_df['publish_time'].dt.hour

# Group by hour and country
hourly_counts = combined_df.groupby(['publish_hour', 'Country']).size().reset_index(name='Count')

# Plot it
plt.figure(figsize=(12, 6))
sns.lineplot(data=hourly_counts, x='publish_hour', y='Count', hue='Country', marker='o')

plt.title('Trending Videos by Hour of Upload (India vs USA)', fontsize=14)
plt.xlabel('Hour of Day (24h Format)')
plt.ylabel('Number of Videos')
plt.xticks(range(0, 24))
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Group by country and category to find average views
views_by_category = combined_df.groupby(['Country', 'Category'])['views'].mean().reset_index()

# Sort by view count
views_by_category = views_by_category.sort_values(by='views', ascending=False)

# Plot
plt.figure(figsize=(14, 7))
sns.barplot(data=views_by_category, x='Category', y='views', hue='Country', palette='Set1')

plt.title('Average Views by Category (India vs USA)', fontsize=14)
plt.ylabel('Average View Count')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Merge all titles into one string for each country
india_titles = " ".join(combined_df[combined_df['Country'] == 'India']['title'])
usa_titles = " ".join(combined_df[combined_df['Country'] == 'USA']['title'])

# Optional: Remove common stopwords
stopwords = set(STOPWORDS)

# Generate word clouds
india_wc = WordCloud(width=800, height=400, background_color='white', stopwords=stopwords).generate(india_titles)
usa_wc = WordCloud(width=800, height=400, background_color='white', stopwords=stopwords).generate(usa_titles)

# Plot them
plt.figure(figsize=(16, 8))

plt.subplot(1, 2, 1)
plt.imshow(india_wc, interpolation='bilinear')
plt.axis('off')
plt.title('India Trending Titles Word Cloud')

plt.subplot(1, 2, 2)
plt.imshow(usa_wc, interpolation='bilinear')
plt.axis('off')
plt.title('USA Trending Titles Word Cloud')

plt.tight_layout()
plt.show()
```
